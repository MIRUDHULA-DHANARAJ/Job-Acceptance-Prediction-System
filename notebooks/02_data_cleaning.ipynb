```python
import pandas as pd
import numpy as np

df=pd.read_csv("../data/raw/HR_Job_Placement_Dataset.csv")
```

```python
cleaning_col=df[['internship_experience','layoff_history','company_tier','job_role_match','relevant_experience','career_switch_willingness','bond_requirement','relocation_willingness','competition_level','status']].columns
```

```python
for col in cleaning_col:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
num_cols=df[['ssc_percentage','hsc_percentage','notice_period_days','employment_gap_months']].columns


```

```python
for col in num_cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
cat_col=df[['career_switch_willingness','relevant_experience','job_role_match','layoff_history','relocation_willingness']].columns
```

```python
for col in cat_col:
    df[col] = df[col].replace('nan', np.nan)
```

```python
for col in cat_col:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
label_en=df[['internship_experience','layoff_history']].columns
```

```python
for col in label_en:
    df[col] = df[col].map({"yes": 1, "no": 0})
```

```python
df['career_switch_willingness']=df['career_switch_willingness'].map({"willing": 1, "not willing": 0})
```

```python
df["job_role_match"]=df["job_role_match"].map({"matched": 1, "not matched": 0})
```

```python
df['bond_requirement']=df['bond_requirement'].map({"required": 1, "not required": 0})
```

```python
df['relevant_experience']=df['relevant_experience'].map({"relevant": 1, "not relevant": 0})
```

```python
df['relocation_willingness']=df['relocation_willingness'].map({"willing": 1, "not willing": 0})
```

```python
df['status']=df['status'].map({"placed": 1, "not placed": 0})
```

```python
df['competition_level']=df['competition_level'].map({"high": 2, "medium": 1, "low": 0})
```

```python
df['company_tier']=df['company_tier'].map({"tier 1": 1, "tier 2": 2, "tier 3": 3})
```

```python
df = pd.get_dummies(df, columns=['degree_specialization'], drop_first=True)
```

```python
df.columns.tolist()
```

```python
deg_spec_col=df[['degree_specialization_Electronics','degree_specialization_Information Technology','degree_specialization_Mechanical','degree_specialization_Others']].columns
```

```python
print(df[deg_spec_col].dtypes)
```

```python
df[deg_spec_col] = df[deg_spec_col].astype(int)
```

```python
df = df.drop('gender', axis=1)
```
