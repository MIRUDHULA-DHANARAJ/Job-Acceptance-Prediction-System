```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../data/cleaned/cleaned_unscaled.csv")
```

```python
df['years_of_experience'].unique()
```

```python
df['experience_category'] = pd.cut(df['years_of_experience'], bins=[-1,0,3,5], labels=['Fresher', 'Junior', 'Senior'],
include_lowest=True)
```

```python
df['academic_avg'] = df[['ssc_percentage','hsc_percentage','degree_percentage']].mean(axis=1)
```

```python
df['academic_performance_bands'] = pd.cut(df['academic_avg'], bins=[0,60,75,100], labels=['Low', 'Average', 'Good'],include_lowest=True)
```

```python
df['skills_match_level']=pd.cut(df['skills_match_percentage'], bins=[0,50,75,100], labels=['Low', 'Average', 'Good'],
                                     include_lowest=True)
```

```python
df['interview_performance_avg']=df[['technical_score','aptitude_score','communication_score']].mean(axis=1)
```

```python
df['interview_performance_category']=pd.cut(df['interview_performance_avg'], bins=[0,50,75,100], labels=['Low', 'Average', 'Good'],include_lowest=True)    
```

```python
df['placement_score_raw'] = (
    df['skills_match_percentage'] * 0.30 +
    df['technical_score'] * 0.25 +
    df['aptitude_score'] * 0.25 +
    df['communication_score'] * 0.20 +
    df['years_of_experience'] * 0.05
)

df['placement_probability'] = (
    (df['placement_score_raw'] - df['placement_score_raw'].min()) /
    (df['placement_score_raw'].max() - df['placement_score_raw'].min())
)
```

```python
print(df[['experience_category',
          'academic_performance_bands',
          'skills_match_level',
          'interview_performance_category',
          'placement_probability']].head())
```
