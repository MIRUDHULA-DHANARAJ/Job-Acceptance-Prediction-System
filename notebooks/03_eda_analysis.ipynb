```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../data/cleaned/cleaned_unscaled.csv")
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='status')
plt.title('Job Acceptance Distribution')
plt.xlabel('Job Acceptance')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='status', y='skills_match_percentage')
plt.title('Skills Match Percentage by Job Acceptance')
plt.xlabel('Job Acceptance')
plt.ylabel('Skills Match Percentage')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='company_tier', hue='status')
plt.title('Company Tier Distribution by Job Acceptance')
plt.xlabel('Company Tier')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='status', y='years_of_experience')
plt.title('Years of Experience by Job Acceptance')
plt.xlabel('Job Acceptance')
plt.ylabel('Years of Experience')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='competition_level', hue='status')
plt.title('Competition Level Distribution by Job Acceptance')
plt.xlabel('Competition Level')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['technical_score', 'aptitude_score', 
            'communication_score', 'skills_match_percentage',
            'years_of_experience', 'ssc_percentage', 
            'hsc_percentage', 'status']].corr(), annot=True, fmt='.2f',cmap='coolwarm')
plt.title('Correlation Heatmap')    
plt.show()
    
```
