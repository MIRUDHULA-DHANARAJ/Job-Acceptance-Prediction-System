```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../data/cleaned/cleaned_unscaled.csv")
```

```python
## Candidate Performance Analysis
## Academic scores vs placement outcome

plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1) 
sns.boxplot(data=df, x='status',y='ssc_percentage',hue='status', palette='Set2', legend=False)
plt.title("SSC Percentage vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("SSC Percentage")

plt.subplot(1, 3, 2)  
sns.boxplot(data=df, x='status',y='hsc_percentage',hue='status', palette='Set2', legend=False)
plt.title("HSC Percentage vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("HSC Percentage")


plt.subplot(1, 3, 3)  
sns.boxplot(data=df, x='status',y='degree_percentage',hue='status', palette='Set2', legend=False)
plt.title("Degree Percentage vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("Degree Percentage")


plt.tight_layout()
plt.show()
```

```python
df['interview_performance_avg'] = df[['technical_score', 'aptitude_score', 'communication_score']].mean(axis=1)
```

```python
## Skills match vs interview performance


plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, 
                x='skills_match_percentage', 
                y='interview_performance_avg',
                hue='status')  
plt.title("Skills Match Score & Interview Performance")     
plt.xlabel("Skills Match Percentage")
plt.ylabel("Interview Performance Average")
plt.tight_layout()
plt.show()
```

```python
## Certification impact on job acceptance

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='certifications_count', 
            y='status',
            palette='Set2',
            order=sorted(df['certifications_count'].unique()))
plt.title("Certifications Count vs Placement Status")   
plt.xlabel("Number of Certifications")
plt.ylabel("Placement Rate")
plt.tight_layout()
plt.show()
```

```python
## Placement & Acceptance Analysis
## Acceptance rate by company tier

plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='company_tier', 
            y='status',
            palette='Set2',
            order=sorted(df['company_tier'].unique()))
plt.title("Company Tier vs Placement Status")
plt.xlabel("Company Tier")
plt.ylabel("Count of Candidates")
plt.tight_layout()
plt.show()

```

```python
## Experience vs placement success

plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='relevant_experience',
            y='status',
            palette='Set2',
            )
plt.title("Relevant Experience vs Placement Status")
plt.xlabel("Relevant Experience")
plt.ylabel("Count of Candidates")
plt.tight_layout()
plt.show()
```

```python
## Interview & Evaluation Analysis
## Interview score vs placement probability

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='status',y='interview_performance_avg',hue='status', palette='Set2', legend=False)
plt.title("Interview Performance vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("Interview Performance Average")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1) 
sns.boxplot(data=df, x='status', y='ssc_percentage', 
            hue='status', palette='Set2', legend=False)
plt.title("SSC Percentage vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("SSC Percentage")

plt.subplot(1, 3, 2) 
sns.boxplot(data=df, x='status', y='hsc_percentage', 
            hue='status', palette='Set2', legend=False)

plt.subplot(1, 3, 3)  
sns.boxplot(data=df, x='status', y='degree_percentage', 
            hue='status', palette='Set2', legend=False)
```
